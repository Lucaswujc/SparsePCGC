```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
idx_file = 0
```

###### ours

```python
csvdir_ours = 'dense_lossy/ours.csv'
df_ours = pd.read_csv(csvdir_ours)
num_rates = 6

filename = df_ours['filedir'][idx_file]
bpp_ours = [df_ours['R'+str(idx)+'_bpp'][idx_file] for idx in range(num_rates)]
d1_ours =[df_ours['R'+str(idx)+'_mseF,PSNR (p2point)'][idx_file] for idx in range(num_rates)]
d2_ours = [df_ours['R'+str(idx)+'_mseF,PSNR (p2plane)'][idx_file] for idx in range(num_rates)]
enc_time_ours = [df_ours['R'+str(idx)+'_enc_time'][idx_file] for idx in range(num_rates)]
dec_time_ours = [df_ours['R'+str(idx)+'_dec_time'][idx_file] for idx in range(num_rates)]

print(filename, 
      '\n', bpp_ours[::-1], 
      '\n', np.round(d1_ours, 2)[::-1], 
      '\n', np.round(d2_ours, 2)[::-1],
     '\n', np.round(enc_time_ours, 2)[::-1],
     '\n', np.round(dec_time_ours, 2)[::-1])
print(np.array(enc_time_ours).mean().round(2))
print(np.array(dec_time_ours).mean().round(2))
```

###### G-PCC

```python
csvdir_gpcc = 'dense_lossy/gpcc.csv'
df_gpcc = pd.read_csv(csvdir_gpcc)
num_rates = 7

filename = df_gpcc['filedir'][idx_file]
bpp_gpcc = [df_gpcc['R'+str(idx)+'_bpp'][idx_file] for idx in range(num_rates)][1:-1]
d1_gpcc =[df_gpcc['R'+str(idx)+'_mseF,PSNR (p2point)'][idx_file] 
          for idx in range(num_rates)][1:-1]
d2_gpcc = [df_gpcc['R'+str(idx)+'_mseF,PSNR (p2plane)'][idx_file] 
           for idx in range(num_rates)][1:-1]

enc_time_gpcc = [df_gpcc['R'+str(idx)+'_enc_time_wall'][idx_file] 
                 for idx in range(num_rates)][1:-1]
dec_time_gpcc = [df_gpcc['R'+str(idx)+'_dec_time_wall'][idx_file] 
                 for idx in range(num_rates)][1:-1]
```

```python
print(filename, '\n', bpp_gpcc, '\n', d1_gpcc, '\n', d2_gpcc)
```

```python
print(filename, 
      '\nbpp_gpcc', bpp_gpcc[::-1], 
      '\nd1_gpcc', np.round(d1_gpcc, 2)[::-1], 
      '\nd2_gpcc', np.round(d2_gpcc, 2)[::-1],
     '\nenc_time_gpcc', np.round(enc_time_gpcc, 2)[::-1],
     '\ndec_time_gpcc', np.round(dec_time_gpcc, 2)[::-1])
print(np.array(enc_time_gpcc).mean().round(2))
print(np.array(dec_time_gpcc).mean().round(2))
```

###### G-PCC (trisoup)

```python
csvdir_trisoup = 'dense_lossy/trisoup.csv'
df_trisoup = pd.read_csv(csvdir_trisoup)
num_rates = 4

filename = df_trisoup['filename'][idx_file]
bpp_trisoup = [df_trisoup['R'+str(idx)+'_bpp'][idx_file] for idx in range(num_rates)][1:]
d1_trisoup = [df_trisoup['R'+str(idx)+'_d1'][idx_file] for idx in range(num_rates)][1:]
d2_trisoup = [df_trisoup['R'+str(idx)+'_d2'][idx_file] for idx in range(num_rates)][1:]
```

```python
print(filename, '\n', bpp_trisoup, '\n', d1_trisoup, '\n', d2_trisoup)
```

###### adl-pcc

```python
csvdir_adl = 'dense_lossy/ADLPCC.csv'
df_adl = pd.read_csv(csvdir_adl)
num_rates = 3

filename = df_adl['filename'][idx_file]
bpp_adl = [df_adl['R'+str(idx)+'_bpp'][idx_file] for idx in range(num_rates)]
d1_adl = [df_adl['R'+str(idx)+'_d1'][idx_file] for idx in range(num_rates)]
d2_adl = [df_adl['R'+str(idx)+'_d2'][idx_file] for idx in range(num_rates)]
```

# plot D1

```python
markersize = 16
linewidth = 4
fig, ax = plt.subplots(figsize=(7.3*2,5*2))

plt.plot(bpp_ours, d1_ours, label='Ours', 
         color='red', linewidth=linewidth, marker='^', markersize=markersize)
plt.plot(bpp_gpcc, d1_gpcc, label='G-PCC', 
         color='green', linewidth=linewidth, marker='o', markersize=markersize)
plt.plot(bpp_trisoup, d1_trisoup, label='G-PCC (T)', 
         color='blue', linewidth=linewidth, marker='d', markersize=markersize)
plt.plot(bpp_pcgcv2, d1_pcgcv2, label='PCGCv2', 
         color='orange', linewidth=linewidth, marker='v', markersize=markersize)
plt.plot(bpp_geocnn, d1_geocnn, label='GeoCNNv2', 
         color='blueviolet', linewidth=linewidth, marker='P', markersize=markersize)
plt.plot(bpp_adl, d1_adl, label='ADL-PCC', 
         color='saddlebrown', linewidth=linewidth, marker='X', markersize=markersize)

# ax.set_xlim([0,bpp_lossless_gpcc+0.05])
# # ax.set_ylim([0.05, 0.15])

ax.locator_params(axis='x', nbins=12)
ax.locator_params(axis='y', nbins=10)
plt.tick_params(labelsize=36)
plt.xlabel("bpp", fontdict={'family':'serif', 'weight':'normal','size':40})
plt.ylabel("D1 PSNR (dB)", fontdict={'family':'serif', 'weight':'normal','size':40})

plt.title(filename, fontdict={'family':'serif', 'weight':'normal','size':60})
plt.grid(ls='-.')
plt.legend(loc='lower right', ncol=2,
           prop={'family':'serif', 'weight':'normal','size':32})
fig.tight_layout()
os.makedirs('figs', exist_ok=True)
fig.savefig( os.path.join('figs', filename+'_d1.pdf'))
```

# plot d2

```python
markersize = 16
linewidth = 4
fig, ax = plt.subplots(figsize=(7.3*2,5*2))

plt.plot(bpp_ours, d2_ours, label='Ours', 
         color='red', linewidth=linewidth, marker='^', markersize=markersize)
plt.plot(bpp_gpcc, d2_gpcc, label='G-PCC', 
         color='green', linewidth=linewidth, marker='o', markersize=markersize)
plt.plot(bpp_trisoup, d2_trisoup, label='G-PCC (T)', 
         color='blue', linewidth=linewidth, marker='d', markersize=markersize)
plt.plot(bpp_pcgcv2, d2_pcgcv2, label='PCGCv2', 
         color='orange', linewidth=linewidth, marker='v', markersize=markersize)
plt.plot(bpp_geocnn, d2_geocnn, label='GeoCNNv2', 
         color='blueviolet', linewidth=linewidth, marker='P', markersize=markersize)
plt.plot(bpp_adl, d2_adl, label='ADL-PCC', 
         color='saddlebrown', linewidth=linewidth, marker='X', markersize=markersize)

# ax.set_xlim([0,bpp_lossless_gpcc+0.05])
# ax.set_ylim([0.05, 0.15])

ax.locator_params(axis='x', nbins=12)
ax.locator_params(axis='y', nbins=10)
plt.tick_params(labelsize=36)
plt.xlabel("bpp", fontdict={'family':'serif', 'weight':'normal','size':40})
plt.ylabel("D2 PSNR (dB)", fontdict={'family':'serif', 'weight':'normal','size':40})

plt.title(filename, fontdict={'family':'serif', 'weight':'normal','size':60})
plt.grid(ls='-.')
plt.legend(loc='lower right', ncol=2,
           prop={'family':'serif', 'weight':'normal','size':32})
fig.tight_layout()
os.makedirs('figs', exist_ok=True)
fig.savefig( os.path.join('figs', filename+'_d2.pdf'))
```
